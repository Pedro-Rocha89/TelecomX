Na coluna 'Churn', principal coluna de anÃ¡lise, existem dados faltantes (''). Portanto, devemos excluÃ­-los

```python
df_final['Churn'].value_counts()
```

```python
df_final = df_final[df_final['Churn'] != '']
df_final.reset_index(drop=True, inplace=True)
df_final
```

```python
df_final['Churn'].value_counts()
```

Outra coluna que apresenta dados vazios (' ') Ã© a Charges.Total. Essa coluna tem uma relaÃ§Ã£o com Charges.Monthly e tenure.

A coluna tenure representa a quantidade de meses em que o cliente estava assinando o serviÃ§o. A coluna Charges.Monthly representa os gastos mensais do cliente e Charges.Total Ã© a quantidade total de gastos, portanto seria uma multiplicaÃ§Ã£o de Charges.Monthly por tenure.

Vamos selecionar todas as linhas onde o tenure = 0, ou seja, os clientes que assinaram o serviÃ§o por 0 meses, e mostrar os resultados para as colunas Charges.Total e Charges.Monthly.

```python
df_final['Charges.Total'].value_counts()
```

```python
df_final.query('tenure == 0')[['Charges.Total', 'Charges.Monthly', 'tenure']]
```

Verificamos que quando tenure = 0, os dados de Charges.Total Ã© vazio (' ').

Vamos agora selecionar os dados onde Charges.Total = ' ', mostrando os resultados para Charges.Monthly e tenure.

```python
df_final[df_final['Charges.Total'] == ' '][['Charges.Total', 'Charges.Monthly', 'tenure']]
```

Percebe-se que todas as linhas de Charges.Total que estÃ£o vazias Ã© porque o cliente nÃ£o assinou nem por um mÃªs. Temos que preencher esse valor com o mesmo valor que estÃ¡ presente em "Charges.Monthly" pois esse representa o total.

```python
idx = df_final[df_final['Charges.Total'] == ' '].index
df_final.loc[idx, "Charges.Total"] = df_final.loc[idx, "Charges.Monthly"]
```

```python
df_final.query('tenure == 0')[['Charges.Total', 'Charges.Monthly', 'tenure']]
```

```python
df_final.info()
```

Por fim, vamos modificar o tipo da variÃ¡vel para float, uma vez que o tipo estava como object.

```python
df_final['Charges.Total'] = df_final['Charges.Total'].astype('float64')
```

```python
df_final.info()
```

#### Criando a Conta DiÃ¡ria

```python
# Criando a coluna "Contas_Diarias"
df_final['Charges.diary'] = df_final['Charges.Monthly']/30
```

```python
df_final
```

#### Padronizando os Dados

```python
# Lista de colunas para mapear 'Yes'/'No' para 1/0
colunas_sim_nao = [
    'Partner',
    'Dependents',
    'PaperlessBilling',
    'Churn'
]

for col in colunas_sim_nao:
    df_final[col] = df_final[col].replace({'Yes': 1, 'No': 0})
```

```python
# Colunas que podem ter 'No internet service' ou 'No phone service'
colunas_servico = [
    'MultipleLines',
    'OnlineSecurity',
    'OnlineBackup',
    'DeviceProtection',
    'TechSupport',
    'StreamingTV',
    'StreamingMovies'
]

for col in colunas_servico:
    df_final[col] = df_final[col].replace({
        'Yes': 1,
        'No': 0,
        'No internet service': 0,
        'No phone service': 0
    })
```

```python
# A coluna 'PhoneService' Ã© binÃ¡ria 'Yes'/'No'
df_final['PhoneService'] = df_final['PhoneService'].replace({'Yes': 1, 'No': 0})
```

```python
# Verificando os tipos e valores Ãºnicos apÃ³s a transformaÃ§Ã£o
print("Tipos de dados apÃ³s a transformaÃ§Ã£o:")
print(df_final[colunas_sim_nao + colunas_servico + ['PhoneService']].dtypes)

print("\nValores Ãºnicos apÃ³s a transformaÃ§Ã£o:")
for col in colunas_sim_nao + colunas_servico + ['PhoneService']:
    print(f"- {col}: {df_final[col].unique()}")
```

## RelatÃ³rio de AnÃ¡lise de EvasÃ£o de Clientes

### 1. IntroduÃ§Ã£o

A **evasÃ£o de clientes**, ou **Churn**, Ã© um desafio central para empresas de telecomunicaÃ§Ãµes. Clientes que cancelam seus serviÃ§os representam uma perda direta de receita e podem indicar insatisfaÃ§Ã£o generalizada. Este relatÃ³rio visa apresentar uma anÃ¡lise exploratÃ³ria de dados detalhada para identificar os **principais fatores e padrÃµes** que levam os clientes a cancelar seus serviÃ§os. O objetivo Ã© fornecer **insights acionÃ¡veis** que possam guiar a empresa na formulaÃ§Ã£o de estratÃ©gias eficazes de retenÃ§Ã£o de clientes.

---
